```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
list_data = ['L01', 'L11', 'L14']

data = {}
for i in list_data:
    value = pd.read_csv('{}.csv'.format(i))
    data[i] = value
    
data['L14'].head()
    
```

```python
for key,value in data.items():
    data[key]['time'] = pd.to_datetime(data[key]['time'])
    data[key].set_index('time', inplace = True)
    
    
data['L14'].head()
```
