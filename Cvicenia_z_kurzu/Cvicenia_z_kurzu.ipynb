```python
# Napíš funkciu is_perfect(number), ktorá pomocou funkcie list_all_divisors() ktorú si už naprogramoval, a 
#zistí, či je číslo dokonalé. To je také že súčet všetkých menších deliteľov ako samotné číslo sa rovná 
#samotnému číslu. Napr. delitele čísla 6 (menšie ako 6) sú 1, 2, 3. Ich súčet je 6. Preto je číslo 6 dokonalé. Funkcia nič nevypisuje ale vracia (pomocou return) True alebo False.


def list_all_divisors(number):
    delitele = [i for i in range(1, number +1) if number % i == 0]
    return delitele

def is_perfect(number2):
    delitele = list_all_divisors(number2)
    sucet = 0
    for i in delitele:
        sucet += i
    if sucet-number2 == number2:
        return True
    else:
        return False

is_perfect(6)
```

```python
# Vytvor funkciu merge(A, B) ktorá spojí 2 slovníky A a B. Konflikty rovnakých kľúčov rieš ťak, že zachováš 
# hodnotu z oboch slovníkov.
#{1 : "one", 2 : ["two", "dva"], 3 : "three"}

def merge(dict1, dict2):
    for key, value in dict2.items():
        if key in dict1 and dict2[key] != dict1[key]:
            dict1[key] = [dict1[key], dict2[key]]
        else:
            dict1.update({key: value})
    return dict1

A = {2 : "dva", 3 : "three"}
B = {1 : "one", 2 : "two", 3: 'three', 4: 'four'}
merge(A, B)


    
    
```

```python
# following turtles by themselves

import turtle
import random
n = 10
turtles = []
turtle.tracer(0)
turtle.bgcolor("black")
for i in range(n):
    t = turtle.Turtle()
    t.pensize(3)
    t.pencolor("#{:06x}".format(random.randrange(256 ** 3)))
    t.pu()
    t.setpos(random.randint(-300, 300),
             random.randint(-300, 300))
    t.pd()
    t.ht()
    turtles.append(t)
while True:
    for i in range(n):
        j = (i + 1) % n  # index nasledovnej
        uhol = turtles[i].towards(turtles[j])
        vzd = turtles[i].distance(turtles[j])
        turtles[i].seth(uhol)
        turtles[i].fd(vzd)
        turtles[i].fd(vzd/10 - vzd)
    turtle.update()
```

```python
import turtle

turtle.delay(1000)
t = turtle.Turtle()
t.speed(0)

def spirala(d):
    print('volanie spirala({})'.format(d))
    if d > 100:
        pass # nerob nic
        print('trivialny pripad, nerobim nic')
    else:
        t.fd(d)
        t.lt(60)
        print('rekurzivne volam spirala({})'.format(d+3))
        spirala(d+3)
        print('rekurzivne volam spirala({})'.format(d+3))
        
spirala(90)
```

```python
import tkinter
canvas = tkinter.Canvas(width=640, height=480)
canvas.pack()

def motion(event):
    print("hybem sa na {},{}".format(event.x, event.y))
def drag(event):
    print("ťahám myšku na {},{}".format(event.x, event.y))
def press(event):
    print("stlačil som na {},{}".format(event.x, event.y))
def release(event):
    print("pustil som na {},{}".format(event.x, event.y))
canvas.bind("<Motion>", motion)
canvas.bind("<B1-Motion>", drag)
canvas.bind("<ButtonPress-1>", press)
canvas.bind("<ButtonRelease-1>", release)

canvas.mainloop()
```

```python
# draw rectangle by dragging mouse

import tkinter
canvas = tkinter.Canvas(width=640, height=480)
canvas.pack()
rect_id = None
def mouse_drag(event):
    x1,y1,x2,y2 = canvas.coords(rect_id)
    canvas.coords(rect_id, x1, y1, event.x, event.y)
def create_rect(event):
    global rect_id
    rect_id = canvas.create_rectangle(event.x, event.y, event.x, event.y)
canvas.bind("<B1-Motion>", mouse_drag)
canvas.bind("<ButtonPress-1>", create_rect)
canvas.mainloop()
```

```python
import tkinter
canvas = tkinter.Canvas(width=640, height=480)
canvas.pack()
def kresli():
    global suradnice
    suradnice += [x, y]
    canvas.coords(line_id, suradnice)
def left(event):
    global x
    x -= 10
    kresli()
def right(event):
    global x
    x += 10
    kresli()
def up(event):
    global y
    y -= 10
    kresli()
def down(event):
    global y
    y += 10
    kresli()
x, y = 100,100
suradnice = [x, y]
line_id = canvas.create_line(0, 0, 0, 0)
canvas.bind_all('<Left>', left)
canvas.bind_all('<Right>', right)
canvas.bind_all('<Up>', up)
canvas.bind_all('<Down>', down)
canvas.mainloop()
```

```python
# Do plochy nakresli tesne vedľa seba 8 štvorcov veľkosti 50x50. Na každé kliknutie sa príslušný kliknutý 
# štvorec zafarbí na červeno (alebo na nejakú náhodnú farbu). Poradové číslo štvorca v rade zistíš z x-ovej 
# súradnice kliknutého bodu event.x // 50. Skontroluj, či aj event.y je v správnom intervale, t. j. kliklo sa 
# do vnútra niektorého štvorca a nie niekde mimo.

import tkinter
import random
canvas = tkinter.Canvas(width=640, height=480)
canvas.pack()

    
def kresli_stvorce(pocet):
    global x,y
    for i in range(0,pocet):
        canvas.create_rectangle(x-25, y-25,x+25, y+25)
        x+=52 

        
def zafarbi(event):
    id_stvorca = (event.x)// 51
    farba = "#{:06x}".format(random.randrange(256 ** 3))
    if int(event.y) in range(75, y+25):
        canvas.itemconfig(id_stvorca, fill = farba)
        
x,y = (75, 100)
kresli_stvorce(8)    
canvas.bind("<ButtonPress-1>", zafarbi)

canvas.mainloop()
```

```python
# V ploche sa nachádza jeden červený štvorček veľkosti 50x50. Keď klikneme do jeho vnútra (počíta sa aj obvod),
# môžeme ho ťahať, teda posúvať po ploche podľa pohybu myši (inak ťahanie s kliknutím mimo štvorček nerobí nič).

import tkinter
import random

width=640
height=480
canvas = tkinter.Canvas(width=width, height=height)
canvas.pack()

x, y = width/2, height/2

rect = canvas.create_rectangle(x-25, y-25,x+25, y+25, fill = 'red')

def mouse_drag(event):   
    global rect
    x1,y1,x2,y2 = canvas.coords(rect)
    if int(event.x) in range(int(x1), int(x2)) and int(event.y) in range(int(y1),int(y2)):
            canvas.coords(rect, event.x-25, event.y-25, event.x+25, event.y+25)

    
canvas.bind("<B1-Motion>", mouse_drag)

canvas.mainloop()
```

```python
# Stláčaním malých a veľkých písmen abecedy sa tieto vypisujú nejakým fontom náhodne na obrazovku. Použi 
# metódu bind_all('<Key>', ...) pričom vo viazanej funkcii pracujte s hodnotou `event.char`

import tkinter
import random
canvas = tkinter.Canvas(width=640, height=480)
canvas.pack()


def klavesa(event):
    x, y = random.randint(0,640), random.randint(0,480)
    canvas.create_text(x, y, text = event.char, font = 'Arial 16', fill = 'blue', angle = 0)
    
canvas.bind_all("<Key>", klavesa)
canvas.mainloop()
```
